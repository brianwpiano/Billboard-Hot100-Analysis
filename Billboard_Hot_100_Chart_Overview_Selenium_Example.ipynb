```python
# Import all libraries
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
import time
from datetime import datetime, timedelta
import plotly.express as px
```

```python
# Lets Check the Error
driver = webdriver.Chrome()

## Perform Scrap on  1 page only

url = f'https://www.billboard.com/charts/hot-100/2024-01-06/'  ## Let's do a specific date for now as it takes a time to process the data
driver.get(url)

driver.implicitly_wait(10) 

w = WebDriverWait(driver, 10)
w.until(EC.presence_of_element_located((By.CLASS_NAME,"c-title")))  ## Lets wait for a specific class of continer to appear

elements = driver.find_elements(By.CLASS_NAME, "o-chart-results-list-row-container") ## Find the common class across the list of music
```

```python
len(elements)
```

```python
title = []
signer = []
last_week_position = []
peak_position = []
weaks_on_chart = []
current_week_position = []

i=0

for element in elements:
    try:
        # Extract titles
        i = i+1
        title1_element = element.find_element(By.XPATH, f'/html/body/div[4]/main/div[2]/div[3]/div/div/div/div[2]/div[{i}]/ul/li[4]/ul/li[1]/h3').text
        
        # Extract Singer
        title2_element = element.find_element(By.XPATH, f'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div[{i}]/ul/li[4]/ul/li[1]/span').text
        title.append(title1_element)
        signer.append(title2_element)

        # This weak position
        current_week_pos = element.find_element(By.XPATH, f'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div[{i}]/ul/li[1]/span').text
        current_week_position.append(current_week_pos)

        # Extract last week
        last_week_pos = element.find_element(By.XPATH, f'/html/body/div[4]/main/div[2]/div[3]/div/div/div/div[2]/div[{i}]/ul/li[4]/ul/li[4]/span').text
        last_week_position.append(last_week_pos)

        # # Peak Pos
        peak_pos = element.find_element(By.XPATH, f'/html/body/div[4]/main/div[2]/div[3]/div/div/div/div[2]/div[{i}]/ul/li[4]/ul/li[5]/span').text
        peak_position.append(peak_pos)

        # # Weaks on Chart
        wks_chart = element.find_element(By.XPATH, f'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div[{i}]/ul/li[4]/ul/li[6]/span').text
        weaks_on_chart.append(wks_chart)

    except:
        pass

# driver.quit()
```

```python
df_songs = pd.DataFrame(title, columns=['Title'])
df_songs['Artist'] = signer
df_songs['Current_Week_Position'] = current_week_position
df_songs['Last_Week_Position'] = last_week_position
df_songs['Peak_Position'] = peak_position
df_songs['Weekly_Chart'] = weaks_on_chart
```
